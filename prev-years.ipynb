```python
import matplotlib.pyplot as plot
import numpy as np
import random          #provides functions for generating random numbers and performing random selections
random.seed(17)

M=100                    #number of iteration in the experiment
N=4000                      #number of coin toss in each iteration
Coin = ['H','T']
y = np.linspace(1,M,M)      #It generates M evenly spaced numbers over the interval from 1 to M in form of an array
#linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None, axis=0)
E=[]

for j in range(M):
    heads = 0
    tails = 0
    for i in range(N):
        x = random.choice(Coin)
        if (x=='H'):
            heads += 1
        else:
            tails -= 1
    z = (heads+tails)
    E.append(z)
plot.plot(y,E)
plot.xlabel('Iteration')
plot.ylabel('Difference (Heads - Tails)')
plot.title('Difference between Heads and Tails in Coin Flips')
# Add description
description = "The resulting plot shows the cumulative score over the course of the coin tosses, indicating how the balance between heads and tails changes over time.".format(num_flips)
plt.text(0.5, -0.18, description, horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes, fontsize=10, wrap=True)
plot.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import random

# Set a seed for reproducibility
random.seed(267)

# Number of coin flips
num_flips = 40

# Possible outcomes of a coin flip
coin_sides = ['Head', 'Tail']

# X-axis values representing the flip number
flip_numbers = np.arange(num_flips + 1)

# Initialize counts and cumulative score
head_count = 0
cumulative_score = 0
scores_over_time = [0]  # List to store cumulative scores

# Simulate coin flips
for flip_num in range(num_flips):
    # Randomly select a side of the coin
    side = random.choice(coin_sides)

    # Update counts and cumulative score
    if side == 'Head':
        head_count += 1
        cumulative_score += 1
    else:
        cumulative_score -= 1

    # Append the cumulative score after each flip
    scores_over_time.append(cumulative_score)

# Plotting
plt.plot(flip_numbers, scores_over_time, marker='o', linestyle='-')
plt.xlabel('Flip Number')
plt.ylabel('Cumulative Score')
plt.title('Cumulative Score Over Coin Flips')
plt.grid(True)  # Add grid for better visualization
plt.tight_layout()  # Adjust layout to prevent clipping of labels
# Add description
description = "The plot visualizes the cumulative score over the tosses, reflecting the progression of 'wins' (heads flipped) against 'losses' (tails flipped).".format(num_flips)
plt.text(0.5, -0.18, description, horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes, fontsize=10, wrap=True)
plt.show()

plt.hist(scores_over_time, bins=30, density=True)
plt.xlabel('Cumulative Score')
plt.ylabel('Probability Density')
plt.title('Histogram of Cumulative Score Over Coin Flips')
description = "As the coin flips are random, the distribution of cumulative scores will typically be centered around 0, reflecting an equal chance of obtaining heads or tails over multiple flips.".format(num_flips)
plt.text(0.5, -0.2, description, horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes, fontsize=10, wrap=True)
plt.show()
```

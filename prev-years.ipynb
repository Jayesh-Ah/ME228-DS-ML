Coin Toss in Python

```python
import matplotlib.pyplot as plot
import numpy as np
import random          #provides functions for generating random numbers and performing random selections

M=100                    #number of iteration in the experiment
N=4000                      #number of coin toss in each iteration
Coin = ['H','T']
y = np.linspace(1,M,M)      #It generates M evenly spaced numbers over the interval from 1 to M in form of an array
E=[]

for j in range(M):
    heads = 0
    tails = 0
    for i in range(N):
        x = random.choice(Coin)
        if (x=='H'):
            heads += 1
        else:
            tails -= 1
    z = (heads+tails)
    E.append(z)
plot.plot(y,E)
plot.show()
```

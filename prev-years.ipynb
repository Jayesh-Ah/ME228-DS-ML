```python
import matplotlib.pyplot as plot
import numpy as np
import random          #provides functions for generating random numbers and performing random selections
random.seed(17)

M=100                    #number of iteration in the experiment
N=4000                      #number of coin toss in each iteration
Coin = ['H','T']
y = np.linspace(1,M,M)      #It generates M evenly spaced numbers over the interval from 1 to M in form of an array
#linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None, axis=0)
E=[]

for j in range(M):
    heads = 0
    tails = 0
    for i in range(N):
        x = random.choice(Coin)
        if (x=='H'):
            heads += 1
        else:
            tails -= 1
    z = (heads+tails)
    E.append(z)
plot.plot(y,E)
plot.xlabel('Iteration')
plot.ylabel('Difference (Heads - Tails)')
plot.title('Difference between Heads and Tails in Coin Flips')
# Add description
description = "The resulting plot shows the cumulative score over the course of the coin tosses, indicating how the balance between heads and tails changes over time.".format(num_flips)
plt.text(0.5, -0.18, description, horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes, fontsize=10, wrap=True)
plot.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import random

# Set a seed for reproducibility
random.seed(267)

# Number of coin flips
num_flips = 40

# Possible outcomes of a coin flip
coin_sides = ['Head', 'Tail']

# X-axis values representing the flip number
flip_numbers = np.arange(num_flips + 1)

# Initialize counts and cumulative score
head_count = 0
cumulative_score = 0
scores_over_time = [0]  # List to store cumulative scores

# Simulate coin flips
for flip_num in range(num_flips):
    # Randomly select a side of the coin
    side = random.choice(coin_sides)

    # Update counts and cumulative score
    if side == 'Head':
        head_count += 1
        cumulative_score += 1
    else:
        cumulative_score -= 1

    # Append the cumulative score after each flip
    scores_over_time.append(cumulative_score)

# Plotting
plt.plot(flip_numbers, scores_over_time, marker='o', linestyle='-')
plt.xlabel('Flip Number')
plt.ylabel('Cumulative Score')
plt.title('Cumulative Score Over Coin Flips')
plt.grid(True)  # Add grid for better visualization
plt.tight_layout()  # Adjust layout to prevent clipping of labels
# Add description
description = "The plot visualizes the cumulative score over the tosses, reflecting the progression of 'wins' (heads flipped) against 'losses' (tails flipped).".format(num_flips)
plt.text(0.5, -0.18, description, horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes, fontsize=10, wrap=True)
plt.show()

plt.hist(scores_over_time, bins=30, density=True)
plt.xlabel('Cumulative Score')
plt.ylabel('Probability Density')
plt.title('Histogram of Cumulative Score Over Coin Flips')
description = "As the coin flips are random, the distribution of cumulative scores will typically be centered around 0, reflecting an equal chance of obtaining heads or tails over multiple flips.".format(num_flips)
plt.text(0.5, -0.2, description, horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes, fontsize=10, wrap=True)
plt.show()
```

Cumulative Distribution Function (CFD) and Density Function (DF)


```python
import matplotlib.pyplot as plt  
import numpy as np 
import math  

# Define the cumulative distribution function (CDF) F(x)
def F(x):
    if x <= 0:
        return 0  
    elif 0 < x <= 1:
        return math.sqrt(x)  
    else:
        return 1  

# Define the probability density function (PDF) f(x)
def f(x):
    if x <= 0:
        return 0  
    elif 0 < x <= 1:
        return 1 / (2 * math.sqrt(x))  
    else:
        return 0  

# Generate an array of x values from -0.5 to 1.5 with a small step size for smooth plotting
x = np.arange(-0.5, 1.5, 0.0001)

# Calculate the values of the PDF for each x and store them in a list df
df = [f(xi) for xi in x]

# Calculate the values of the CDF for each x and store them in a list cdf
cdf = [F(xi) for xi in x]

# Plotting the PDF and CDF with improved appearance
plt.figure(figsize=(8, 6))  # Set figure size for better visualization

# Plot PDF and CDF with customized line styles and colors
plt.plot(x, df, linestyle='-', color='b', linewidth=2, label='PDF: f(x)')  # Plot PDF with blue color
plt.plot(x, cdf, linestyle='--', color='r', linewidth=2, label='CDF: F(x)')  # Plot CDF with red dashed line

plt.ylim([0, 1.1])  # Set y-axis limits
plt.xlabel('x')  # Label for x-axis
plt.ylabel('Probability Density')  # Label for y-axis
plt.title('Probability Density Function (PDF) and Cumulative Distribution Function (CDF)')  # Title of the plot

plt.legend(loc='upper right')  # Add legend at upper right corner
plt.grid(True, linestyle='--', alpha=0.6)  # Add grid lines with dashed style and transparency

plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()  # Display the plot
```

```python
import numpy as np

def shared_birthday_probability(n):
    # Calculate the probability of no shared birthday
    prob_no_shared_birthday = np.prod((365 - np.arange(n)) / 365)
    
    # Calculate the probability of at least one shared birthday
    prob_shared_birthday = 1 - prob_no_shared_birthday
    
    return prob_shared_birthday

def find_class_size():
    n = 2  # Starting with two people
    while shared_birthday_probability(n) <= 0.5:
        n += 1
    return n

class_size = find_class_size()
print("The class size should be at least", class_size, "for the probability of two students sharing the same birthday to be more than 50%.")
```

MONTY HALL PROBLEM


```python
import numpy as np

def monty_hall(iteration):
    switch_wins = 0
    stick_wins = 0
    
    for n in range(iteration):
        car_door = np.random.randint(3)
        initial_choice = np.random.randint(3)
        
        remaining_doors = [door for door in range(3) if door != initial_choice and door != car_door]
        
        open_door = np.random.choice(remaining_doors)
        
        switch_choice = list(range(3))
        switch_choice.remove(open_door)
        switch_choice.remove(initial_choice)
        
        if switch_choice == car_door:
            switch_wins += 1
        if initial_choice == car_door:
            stick_wins += 1
        
    switch_win_percentage = (switch_wins / iteration) * 100
    stick_win_percentage = (stick_wins / iteration) * 100

    print("Switching wins {:.2f}% of the time.".format(switch_win_percentage))
    print("Sticking wins {:.2f}% of the time.".format(stick_win_percentage))

# Perform the simulation with a large number of trials
monty_hall(100000)
            
```

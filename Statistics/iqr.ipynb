```python
# Outliers

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset = [2, 4, 7, 7, 3, 6, 1, 8, 23, 43, 64, 21, 12, 21, 32, 75, 58, 87, 96, 34, 642, 2, 5, 32, 16, 83, 92, 42, 65, 87, 23, 45, 79, 85, 23, 1034]
```

```python
np.mean(dataset)
```

```python
np.std(dataset)
```

```python
outliers = []

def detect_outliers(data):
    threshold = 3       # 3rd Standard Deviation
    mean = np.mean(data)
    std = np.std(data)
    
    for i in data:
        z_score = (i-mean)/std
        if np.abs(z_score) > threshold:
            outliers.append(i)
    return outliers
```

```python
detect_outliers(dataset)
```

```python
sns.boxplot(dataset, orient='v')
```

```python
sns.histplot(dataset, kde=True)
```

```python
dataset = sorted(dataset)            
```

```python
q1, q3 = np.percentile(dataset, [25, 75])           # Calculate 1st quartile and 3rd quartile
print(q1, q3)
```

```python
iqr = q3 - q1       # Calc inter quartile range
print(iqr)
```

```python
lower_fence = q1 - (1.5*iqr)
higher_fence = q3 + (1.5*iqr)

print(lower_fence, higher_fence)
```
